# COVID-19 Data Analysis based on the Tokyo 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import time
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
from IPython.display import display
import requests
```

```python
"""
get file from open data source 
https://catalog.data.metro.tokyo.lg.jp/dataset/t000010d0000000068/resource/c2d997db-1450-43fa-8037-ebb11ec28d4c

"""

URL = 'https://stopcovid19.metro.tokyo.lg.jp/data/130001_tokyo_covid19_patients.csv'
receiveData = requests.get(URL, allow_redirects=True)

open('RawData/covid-19-infectedPersonData.csv', 'wb').write(receiveData.content)
```

```python
dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d')
df = pd.read_csv('RawData/covid-19-infectedPersonData.csv', sep=',', parse_dates=['公表_年月日'])
df = df.fillna(0)
df = df.replace('男性','M')
df = df.replace('女性','F')
df = df.replace('代', '')
df = df.replace('不明', '-1')
df = df.replace('調査中', '-2')

for i in range(len(df.index)):
    buf = df.iloc[i,8]
    id = buf.find('代')
   
    if(id > 0):
        buf = buf[:id]
        df.iloc[i,8] = buf 
        
    id = buf.find('歳未満')
   
    if(id > 0):
        buf = '0'
        df.iloc[i,8] = buf 

    
df.iloc[:,8] = df.iloc[:,8].astype('float')    
display(df)
display(df.dtypes)
```

```python
def plotFitting(df, plotStart = "2020-03", plotEND = "2020-05"):

    def func(x, a,b,c):
        return a*np.exp(b*x)+c
    #a*pow(x,2) + b*pow(x,1)+ c

    dates = df.iloc[:,4].drop_duplicates()
    dates = dates.reset_index(drop = True)  
    
    initalDayID = dates[dates >= datetime.datetime.strptime(plotStart, '%Y-%m')].index[0]
    endDayID = dates[dates <= datetime.datetime.strptime(plotEND, '%Y-%m')].index[-1]
    
    #display(dates)

    array = np.array([])

    for date in dates[initalDayID:endDayID]:
        df_tmp = df[df['公表_年月日'] == date]
        td = date-dates[initalDayID]   

        if len(array) == 0:
            array = np.array([[date, td.days, len(df_tmp.index)]])
        else:
            array = np.append(array, np.array([[date, td.days, len(df_tmp.index)]]), axis= 0)        
    #display(array)

    x = array[:,0]
    y = array[:,2].astype('float')
    xdata = array[:,1].astype('float')

    popt, pcov = curve_fit(func, xdata, y)

    display(popt)


    ydata = func(xdata, *popt)
    R2 = r2_score(y, ydata)

    plt.style.use('bmh')
    plt.xticks(rotation=45)
    plt.xlabel('date')
    plt.ylabel('infected peoples')
    #plt.ylim([0,200])

    plt.bar(x,y, label = "Daily report")

    x = np.arange(plotStart, plotEND, dtype='datetime64[D]')
    xdata=np.array(range(len(x))).astype("float")
    plt.plot(x, func(xdata, *popt), ls = '--', c = 'red' ,
                label='fit: A * e^Bx +C\nA=%3.2f, B=%3.2f, C=%3.2f, \nR^2 = %2.3f' % (popt[0],popt[1],popt[2], R2))
    plt.legend(loc = 'best')
    plt.show()
```

```python
plotFitting(df)
```
